### Convert 'Timestamp' to DateTime format

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

### Check for missing values

```python
missing_values = df.isnull().sum().sort_values(ascending = False)
```

### Visualize Missing Values

```python
plt.figure(figsize=(16,8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Data Heatmap", fontsize = 16)
plt.show()
```

### Descriptive Statistics for numerical columns 

```python
print("Numerical Summary:\n", df.describe())
```

### Descriptive Statistics for Categorical columns

```python
print("\n Categorical Summary:\n", df.describe(include = 'object'))
```

### Distribution of Packet Length and Anomaly Scores

```python
fig, axes = plt.subplots(1, 2, figsize = (14,5))

sns.histplot(df['Packet Length'], bins=30, kde = True, color = 'skyblue', ax=axes[0])
axes[0].set_title('Distribution of Packet Length', fontsize = 14)
axes[0].set_xlabel('Packet Length')

sns.histplot(df['Anomaly Scores'], bins=30, kde = True, color = 'orange', ax=axes[1])
axes[1].set_title('Distribution of Anomaly Scores', fontsize = 14)
axes[1].set_xlabel('Anomaly Scores')

plt.tight_layout()
plt.show()
```

### Counrt of Severity Levels

```python
plt.figure(figsize =(8,5))
sns.countplot(y = 'Severity Level', data=df, palette= 'viridis', order=df['Severity Level'].value_counts().index)
plt.title('Count of Severity Levels', fontsize = 16)
plt.xlabel('Count')
plt.ylabel('Severity Level')
plt.show()
```

### Attack Types by Traffic Type

```python
plt.figure(figsize=(10,6))
sns.countplot(y = 'Attack Type', hue = 'Traffic Type', data=df, palette = 'Set2')
plt.title('Attack Types by Traffic Type', fontsize = 16)
plt.xlabel('Count')
plt.ylabel('Attack Type')
plt.legend(title = 'Traffic Type')
plt.show()
```

### Attack Trends Over Time 

```python
df['Month-Year'] = df['Timestamp'].dt.to_period('M')
plt.figure(figsize = (12,6))
attack_trends = df['Month-Year'].value_counts().sort_index()
attack_trends.plot(kind = 'line', marker = 'o', color = 'teal')
plt.title('Attack Trends Over Time', fontsize = 16)
plt.xlabel('Month-Year')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.show()
```

```python
# Boxplot: Anomaly Scores by Severity Level 

plt.figure(figsize=(10,6))
sns.boxplot(x='Severity Level', y='Anomaly Scores', data=df, palette='coolwarm')
plt.title('Anomaly Scores by Severity Level', fontsize=16)
plt.xlabel('Severity Level')
plt.ylabel('Anomaly Scores')
plt.show()
```

```python
# Top 10 Source IPs for High-Severity Attacks

high_severity_ips = df[df['Severity Level'] == 'High']['Source IP Address'].value_counts().head(10)
plt.figure(figsize=(10,8))
high_severity_ips.plot(kind='bar', color='salmon')
plt.title('Top 10 Source IPs for High-Severity Attacks', fontsize=16)
plt.xlabel('Source IP Address')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

```

```python
# Correlation heatmap for numerical features

numerical_cols = ['Source Port', 'Destination Port', 'Packet Length', 'Anomaly Scores']
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap', fontsize = 16)
plt.show()
```
